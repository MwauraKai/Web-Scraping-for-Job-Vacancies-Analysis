```python
import pandas as pd
import random

# Create synthetic data
job_titles = ["Data Analyst", "Software Engineer", "Marketing Specialist", "UX Designer", "Financial Analyst"]
companies = ["Company A", "Company B", "Company C", "Company D", "Company E"]
locations = ["New York", "San Francisco", "London", "Berlin", "Singapore"]
salaries = [50000, 60000, 70000, 80000, 90000]
ratings = [3.5, 4.0, 4.5, 3.0, 5.0]

# Generate random job postings
num_postings = 100
job_postings = []

for _ in range(num_postings):
    posting = {
        "Job Title": random.choice(job_titles),
        "Company": random.choice(companies),
        "Location": random.choice(locations),
        "Salary": random.choice(salaries),
        "Rating": random.choice(ratings)
    }
    job_postings.append(posting)

# Create a DataFrame
df = pd.DataFrame(job_postings)

# Display the dataset
df.head()
```

```python
# Display basic statistics of the dataset
#Run the cell with basic statistics: 
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Salary
plt.figure(figsize=(12, 6))
sns.histplot(df['Salary'], bins=20, kde=True)
plt.title('Distribution of Salary')
plt.show()
#The histogram shows the distribution of salaries in the dataset. Each bar represents a salary range (bin), and the height of the bar indicates the frequency or count of job postings within that salary range.

For example, if you see a tall bar around $70,000, it means there are many job postings with salaries close to $70,000. The KDE plot (the smooth line) provides a smoothed representation of the distribution.
```

```python
# Distribution of Ratings
plt.figure(figsize=(12, 6))
sns.histplot(df['Rating'], bins=20, kde=True)
plt.title('Distribution of Ratings')
plt.show()
```

```python
# Scatter plot of Salary vs Rating
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Salary', y='Rating', data=df)
plt.title('Salary vs Rating')
plt.show()
```

it suggests that there is no clear relationship or correlation between the 'Salary' and 'Rating' variables in your synthetic job posting dataset. In other words, changes in salary are not associated with consistent changes in ratings.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distribution of job titles
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Job Title', palette='viridis')
plt.title('Distribution of Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Number of Postings')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```

```python
# Distribution of Companies
plt.figure(figsize=(12, 6))
sns.countplot(x='Company', data=df, palette='Set2')
plt.title('Distribution of Companies')
plt.xticks(rotation=45, ha='right')
plt.show()
# creates a bar plot to visualize the distribution of companies in the dataset. It helps us understand which companies have more job postings in the dataset.
```

```python
# Distribution of Locations
plt.figure(figsize=(12, 6))
sns.countplot(x='Location', data=df, hue='Location', palette='plasma', legend=False)
plt.title('Distribution of Locations')
plt.xticks(rotation=45, ha='right')
plt.show()

```
